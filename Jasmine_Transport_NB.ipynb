```python
ref = db.reference("Transport_Data")

while True: # Main menu loop
    print("\n===== TRANSPORT DATABASE MENU =====")
    print("1. Display Transport Data")
    print("2. Add Trip")
    print("3. Update Trip")
    print("4. Delete Trip")
    print("5. Features")
    print("6. Exit")

    choice = input("Enter choice: ")

    if choice == "6":
        print("\nExiting program.")
        break # Exit main menu loop

    # DISPLAY
    elif choice == "1":
        Transport_Data = ref.get()
        print("\nTrip List:")
        if Transport_Data:
            for key, trip_item in Transport_Data.items():
                print(f"Trip ID: {trip_item['trip_id']}, Route ID: {trip_item['route_id']}, Origin: {trip_item['origin']}, Destination: {trip_item['destination']}, Date: {trip_item['date']}, Departure Time: {trip_item['departure_time']}, Arrival Time: {trip_item['arrival_time']}, Travel Duration (min): {trip_item['travel_duration_min']}, Bus Operator: {trip_item['bus_operator']}, Bus Type: {trip_item['bus_type']}, Capacity: {trip_item['capacity']}, Passengers: {trip_item['passengers']}, Occupancy Rate: {trip_item['occupancy_rate']}, Fare (Regular): {trip_item['fare']['regular']}, Fare (Student): {trip_item['fare']['student']}, Fare (Senior): {trip_item['fare']['senior']}, Status: {trip_item['status']}")
        else:
            print("No transport data available.")

    # ADD
    elif choice == "2":
        Trip_ID = int(input("Enter trip ID: "))
        Route_ID = int(input("Enter route ID: "))
        Origin = input("Enter origin: ")
        Destination = input("Enter destination: ")
        Date = input("Enter date (YYYY-MM-DD): ")
        Departure_Time = input("Enter departure time: ")
        Arrival_Time = input("Enter arrival time: ")
        Travel_Duration_Min = int(input("Enter travel duration (min): "))
        Bus_Operator = input("Enter bus operator: ")
        Bus_Type = input("Enter bus type: ")
        Capacity = int(input("Enter bus capacity: "))
        Passengers = int(input("Enter the number of passengers: "))
        Occupancy_Rate = float(input("Enter the occupancy rate: "))
        Fare_Regular = int(input("Enter the fare for regular passengers: "))
        Fare_Student = int(input("Enter the fare for student passengers: "))
        Fare_Senior = int(input("Enter the fare for senior passengers: "))
        Status = input("Enter status: ")


        trip_new = {
            "trip_id": Trip_ID,
            "route_id": Route_ID,
            "origin": Origin,
            "destination": Destination,
            "date": Date,
            "departure_time": Departure_Time,
            "arrival_time": Arrival_Time,
            "travel_duration_min": Travel_Duration_Min,
            "bus_operator": Bus_Operator,
            "bus_type": Bus_Type,
            "capacity": Capacity,
            "passengers": Passengers,
            "occupancy_rate": Occupancy_Rate,
            "fare": {
                "regular": Fare_Regular,
                "student": Fare_Student,
                "senior": Fare_Senior},
            "status": Status
            }

        ref.child(str(Trip_ID)).set(trip_new)
        print("Trip added successfully!")

     # UPDATE
    elif choice == "3":
        Trip_ID = input("Enter ID of trip to update: ")
        trip_ref = ref.child(Trip_ID)

        trip_to_update = trip_ref.get()

        if trip_to_update:
            Trip_ID = int(input("Enter new trip ID: "))
            Route_ID = int(input("Enter new route ID: "))
            Origin = input("Enter new origin: ")
            Destination = input("Enter new destination: ")
            Date = input("Enter new date (YYYY-MM-DD): ")
            Departure_Time = input("Enter new departure time: ")
            Arrival_Time = input("Enter new arrival time: ")
            Travel_Duration_Min = int(input("Enter new travel duration (min): "))
            Bus_Operator = input("Enter new bus operator: ")
            Bus_Type = input("Enter new bus type: ")
            Capacity = int(input("Enter new bus capacity: "))
            Passengers = int(input("Enter the new number of passengers: "))
            Occupancy_Rate = float(input("Enter the new occupancy rate: "))
            Fare_Regular = int(input("Enter the new fare for regular passengers: "))
            Fare_Student = int(input("Enter the new fare for student passengers: "))
            Fare_Senior = int(input("Enter the new fare for senior passengers: "))
            Status = input("Enter new status: ")

            trip_ref.update({
                "trip_id": Trip_ID,
                "route_id": Route_ID,
                "origin": Origin,
                "destination": Destination,
                "date": Date,
                "departure_time": Departure_Time,
                "arrival_time": Arrival_Time,
                "travel_duration_min": Travel_Duration_Min,
                "bus_operator": Bus_Operator,
                "bus_type": Bus_Type,
                "capacity": Capacity,
                "passengers": Passengers,
                "occupancy_rate": Occupancy_Rate,
                "fare": {
                    "regular": Fare_Regular,
                    "student": Fare_Student,
                    "senior": Fare_Senior},
                "status": Status
                })

            print("Trip updated successfully!")
        else:
            print("Trip not found.")

    # DELETE
    elif choice == "4":
        Trip_ID = input("Enter ID to delete: ")

        if ref.child(Trip_ID).get():
            ref.child(Trip_ID).delete()
            print("Trip deleted successfully!")
        else:
            print("Trip not found.")

    # FEATURES
    elif choice == "5":
        while True:
            print("\n===== FEATURES MENU =====")
            print("1. Summarize and display a bus operator's trips' status")
            print("2. Summarize and display trip data for a destination")
            print("3. Display schedules for a city")
            print("4. Display fares for a trip")
            print("5. Display bus details of a trip")
            print("6. Exit")

            feature_choice = input("Enter choice: ")

            if feature_choice == "6":
                break

            current_transport_data_dict = ref.get()
            current_transport_list = []
            if current_transport_data_dict:
                current_transport_list = list(current_transport_data_dict.values())


            # FEATURE 1:
            if feature_choice == "1":
                while True:
                    bus_operator_input = input("Enter bus operator (0 to exit): ")
                    if bus_operator_input == "0":
                        break

                    found_operator = False
                    for trip_item in current_transport_list:
                        if trip_item and "bus_operator" in trip_item and trip_item["bus_operator"] == bus_operator_input:
                            found_operator = True
                            break

                    if found_operator:
                        operator_upper = bus_operator_input.upper()
                        print(f"\n===== TRIP STATUS OF {operator_upper} =====")
                        for trip_item in current_transport_list:
                            if trip_item and "bus_operator" in trip_item and trip_item["bus_operator"] == bus_operator_input:
                                print(f"Trip {trip_item['trip_id']}: {trip_item['status']}")
                    else:
                        print("Bus operator not found.")

            # FEATURE 2:
            elif feature_choice == "2":
                while True:
                  destination_input = input("\nEnter destination (0 to exit): ")
                  if destination_input == "0":
                      break

                  found_destination = False
                  for trip_item in current_transport_list:
                      if trip_item and "destination" in trip_item and trip_item["destination"] == destination_input:
                          found_destination = True
                          break

                  if found_destination:
                      destination_upper = destination_input.upper()
                      print(f"\n===== TRIP DATA FOR {destination_upper} =====")
                      for trip_item in current_transport_list:
                          if trip_item and "destination" in trip_item and trip_item["destination"] == destination_input:
                                print(f"""
Trip ID: {trip_item['trip_id']}
Route ID: {trip_item['route_id']}
Origin: {trip_item['origin']}
Destination: {trip_item['destination']}
Date: {trip_item['date']}
Departure Time: {trip_item['departure_time']}
Arrival Time: {trip_item['arrival_time']}
Travel Duration (min): {trip_item['travel_duration_min']}
Bus Operator: {trip_item['bus_operator']}
Bus Type: {trip_item['bus_type']}
Capacity: {trip_item['capacity']}
Passengers: {trip_item['passengers']}
Occupancy Rate: {trip_item['occupancy_rate']}
Fare (Regular): {trip_item['fare']['regular']}
Fare (Student): {trip_item['fare']['student']}
Fare (Senior): {trip_item['fare']['senior']}
Status: {trip_item['status']}""")
                  else:
                      print("Destination not found.")

            # FEATURE 3:
            elif feature_choice == "3":
                while True:
                    city_input = input("\nEnter city (origin or destination) to display schedules (0 to exit): ")
                    if city_input == "0":
                        break

                    found_city = False
                    print(f"\n===== SCHEDULES FOR {city_input.upper()} =====")
                    for trip_item in current_transport_list:
                        if trip_item and ("origin" in trip_item and trip_item["origin"] == city_input or "destination" in trip_item and trip_item["destination"] == city_input):
                            found_city = True
                            print(f"""
Trip ID: {trip_item['trip_id']}
Origin: {trip_item['origin']}
Destination: {trip_item['destination']}
Date: {trip_item['date']}
Departure Time: {trip_item['departure_time']}
Arrival Time: {trip_item['arrival_time']}
Duration: {trip_item['travel_duration_min']}
Bus Operator: {trip_item['bus_operator']}
Status: {trip_item['status']}""")
                    if not found_city:
                        print("No trips found for this city.")

            # FEATURE 4:
            elif feature_choice == "4":
                while True:
                    trip_id_input = input("\nEnter Trip ID to display fares (0 to exit): ")
                    if trip_id_input == "0":
                        break

                    found_trip = False
                    for trip_item in current_transport_list:
                        if trip_item and "trip_id" in trip_item and str(trip_item["trip_id"]) == trip_id_input:
                            found_trip = True
                            print(f"\n===== FARES FOR TRIP {trip_item['trip_id']} =====")
                            print(f"Regular Fare: {trip_item['fare']['regular']} pesos")
                            print(f"Student Fare: {trip_item['fare']['student']} pesos")
                            print(f"Senior Fare: {trip_item['fare']['senior']} pesos")
                            break
                    if not found_trip:
                        print("Trip ID not found.")

            # FEATURE 5:
            elif feature_choice == "5":
                while True:
                    trip_id_input = input("\nEnter Trip ID to display bus details (0 to exit): ")
                    if trip_id_input == "0":
                        break

                    found_trip = False
                    for trip_item in current_transport_list:
                        if trip_item and "trip_id" in trip_item and str(trip_item["trip_id"]) == trip_id_input:
                            found_trip = True
                            print(f"\n===== BUS DETAILS FOR TRIP {trip_item['trip_id']} =====")
                            print(f"Bus Operator: {trip_item['bus_operator']}")
                            print(f"Bus Type: {trip_item['bus_type']}")
                            print(f"Capacity: {trip_item['capacity']} seats")
                            print(f"Passengers: {trip_item['passengers']} passengers")
                            print(f"Occupancy Rate: {trip_item['occupancy_rate']}%")
                            print(f"Status: {trip_item['status']}")
                            break
                    if not found_trip:
                        print("Trip ID not found.")

            else:
                print("Invalid choice. Please enter a number from 1 to 6.")
    else:
        print("Invalid choice. Please enter a number from 1 to 6.")
```
