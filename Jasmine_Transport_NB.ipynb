```python
ref = db.reference("Transport_Data")

while True:
    print("\n===== TRANSPORT DATABASE MENU =====")
    print("1. Display Transport Data")
    print("2. Add Trip")
    print("3. Update Trip")
    print("4. Delete Trip")
    print("5. Exit")

    choice = input("Enter choice: ")

    # DISPLAY
    if choice == "1":
        Transport_Data = ref.get()
        print("\nTrip List:")
        if Transport_Data:
            for key, trip in Transport_Data.items():
                print(f"Trip ID: {trip['trip_id']}, Route ID: {trip['route_id']}, Origin: {trip['origin']}, Destination: {trip['destination']}, Date: {trip['date']}, Departure Time: {trip['departure_time']}, Arrival Time: {trip['arrival_time']}, Travel Duration (min): {trip['travel_duration_min']}, Bus Operator: {trip['bus_operator']}, Bus Type: {trip['bus_type']}, Capacity: {trip['capacity']}, Passengers: {trip['passengers']}, Occupancy Rate: {trip['occupancy_rate']}, Fare (Regular): {trip['fare']['regular']}, Fare (Student): {trip['fare']['student']}, Fare (Senior): {trip['fare']['senior']}, Status: {trip['status']}")
        else:
            print("No transport data available.")

    # ADD
    elif choice == "2":
        Trip_ID = int(input("Enter trip ID: "))
        Route_ID = int(input("Enter route ID: "))
        Origin = input("Enter origin: ")
        Destination = input("Enter destination: ")
        Date = input("Enter date (YYYY-MM-DD): ")
        Departure_Time = input("Enter departure time: ")
        Arrival_Time = input("Enter arrival time: ")
        Travel_Duration_Min = int(input("Enter travel duration (min): "))
        Bus_Operator = input("Enter bus operator: ")
        Bus_Type = input("Enter bus type: ")
        Capacity = int(input("Enter bus capacity: "))
        Passengers = int(input("Enter the number of passengers: "))
        Occupancy_Rate = float(input("Enter the occupancy rate: "))
        Fare_Regular = int(input("Enter the fare for regular passengers: "))
        Fare_Student = int(input("Enter the fare for student passengers: "))
        Fare_Senior = int(input("Enter the fare for senior passengers: "))
        Status = input("Enter status: ")


        trip = {
            "trip_id": Trip_ID,
            "route_id": Route_ID,
            "origin": Origin,
            "destination": Destination,
            "date": Date,
            "departure_time": Departure_Time,
            "arrival_time": Arrival_Time,
            "travel_duration_min": Travel_Duration_Min,
            "bus_operator": Bus_Operator,
            "bus_type": Bus_Type,
            "capacity": Capacity,
            "passengers": Passengers,
            "occupancy_rate": Occupancy_Rate,
            "fare": {
                "regular": Fare_Regular,
                "student": Fare_Student,
                "senior": Fare_Senior},
            "status": Status
            }

        ref.child(str(Trip_ID)).set(trip)
        print("Trip added successfully!")

     # UPDATE
    elif choice == "3":
        Trip_ID = input("Enter ID of trip to update: ")
        trip_ref = ref.child(Trip_ID)

        trip = trip_ref.get()

        if trip:
            Trip_ID = int(input("Enter trip ID: "))
            Route_ID = int(input("Enter route ID: "))
            Origin = input("Enter origin: ")
            Destination = input("Enter destination: ")
            Date = input("Enter date (YYYY-MM-DD): ")
            Departure_Time = input("Enter departure time: ")
            Arrival_Time = input("Enter arrival time: ")
            Travel_Duration_Min = int(input("Enter travel duration (min): "))
            Bus_Operator = input("Enter bus operator: ")
            Bus_Type = input("Enter bus type: ")
            Capacity = int(input("Enter bus capacity: "))
            Passengers = int(input("Enter the number of passengers: "))
            Occupancy_Rate = float(input("Enter the occupancy rate: "))
            Fare_Regular = int(input("Enter the fare for regular passengers: "))
            Fare_Student = int(input("Enter the fare for student passengers: "))
            Fare_Senior = int(input("Enter the fare for senior passengers: "))
            Status = input("Enter status: ")

            trip_ref.update({
                "trip_id": Trip_ID,
                "route_id": Route_ID,
                "origin": Origin,
                "destination": Destination,
                "date": Date,
                "departure_time": Departure_Time,
                "arrival_time": Arrival_Time,
                "travel_duration_min": Travel_Duration_Min,
                "bus_operator": Bus_Operator,
                "bus_type": Bus_Type,
                "capacity": Capacity,
                "passengers": Passengers,
                "occupancy_rate": Occupancy_Rate,
                "fare": {
                    "regular": Fare_Regular,
                    "student": Fare_Student,
                    "senior": Fare_Senior},
                "status": Status
                })

            print("Trip updated successfully!")
        else:
            print("Trip not found.")

    # DELETE
    elif choice == "4":
        Trip_ID = input("Enter ID to delete: ")

        if ref.child(Trip_ID).get():
            ref.child(Trip_ID).delete()
            print("Trip deleted successfully!")
        else:
            print("Trip not found.")

    # EXIT
    elif choice == "5":
        print("Exiting program...")
        break

    else:
        print("Invalid choice. Try again.")

"""
Loop → keeps menu running

Conditionals (if-elif-else) → handles choices
CRUD operations:
Create → set()
Read → get()
Update → update()
Delete → delete()
Dictionary (JSON) → data structure
Firebase interaction
"""
```
